## Sample Line 
```python
import matplotlib.pyplot as plt

# Sample data
x = [0, 1, 2, 3, 4, 5]
y = [num ** 2 for num in x]  # y = x^2

# Create the plot
plt.plot(x, y, marker='o', linestyle='-')

# Labeling the axes and title
plt.xlabel('X values')
plt.ylabel('Y values (x squared)')
plt.title('Simple Line Plot')

# Display the plot
plt.show()
```

## Sample Shear and Moment 
```python
import numpy as np

# Beam and load parameters
L = 10.0  # length of the beam (meters)
w = 2.0   # uniform load (kN/m)

# Discretize the beam
x = np.linspace(0, L, 100)

# Shear force: V(x)
V = (w * L / 2) - (w * x)

# Bending moment: M(x)
M = (w * L / 2) * x - (w * x**2) / 2

# Create subplots (2 rows, 1 column)
fig, (ax_shear, ax_moment) = plt.subplots(2, 1, figsize=(8, 6), sharex=True)

# --- Plot the Shear Diagram ---
ax_shear.plot(x, V, label='Shear Force', color='teal')
ax_shear.axhline(0, color='black', linewidth=0.8)  # reference line at 0
ax_shear.set_ylabel('Shear Force V (kN)')
ax_shear.set_title('Shear and Moment Diagrams for a Simply Supported Beam')
ax_shear.legend(loc='best')
ax_shear.grid(True)

# --- Plot the Moment Diagram ---
ax_moment.plot(x, M, label='Bending Moment', color='darkorange')
ax_moment.axhline(0, color='black', linewidth=0.8)  # reference line at 0
ax_moment.set_xlabel('Position along the beam (m)')
ax_moment.set_ylabel('Bending Moment M (kNÂ·m)')
ax_moment.legend(loc='best')
ax_moment.grid(True)

# Adjust layout and display
plt.tight_layout()
plt.show()
```

````python
import numpy as np

# Beam and load parameters
L = 10.0  # length of the beam (meters)
w = 2.0   # uniform load (kN/m)

# Discretize the beam
x = np.linspace(0, L, 100)

# Shear force: V(x)
V = (w * L / 2) - (w * x)

# Bending moment: M(x)
M = (w * L / 2) * x - (w * x**2) / 2

# Create subplots (2 rows, 1 column)
fig, (ax_shear, ax_moment) = plt.subplots(2, 1, figsize=(8, 6), sharex=True)

# --- Plot the Shear Diagram ---
ax_shear.plot(x, V, label='Shear Force', color='teal')
ax_shear.axhline(0, color='black', linewidth=0.8)  # reference line at 0
ax_shear.set_ylabel('Shear Force V (kN)')
ax_shear.set_title('Shear and Moment Diagrams for a Simply Supported Beam')
ax_shear.legend(loc='best')
ax_shear.grid(True)

# --- Plot the Moment Diagram ---
ax_moment.plot(x, M, label='Bending Moment', color='darkorange')
ax_moment.axhline(0, color='black', linewidth=0.8)  # reference line at 0
ax_moment.set_xlabel('Position along the beam (m)')
ax_moment.set_ylabel('Bending Moment M (kNÂ·m)')
ax_moment.legend(loc='best')
ax_moment.grid(True)

# Adjust layout and display
plt.tight_layout()
plt.show()
````
