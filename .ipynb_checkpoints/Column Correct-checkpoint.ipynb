```python
import math
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# ======================
# Given / Assumed Values
# ======================
fc = 30    # Concrete compressive strength, MPa
fy = 415   # Steel yield strength, MPa
b  = 500   # Column width, mm
h  = 500   # Column depth, mm
dbar = 25  # Diameter of main rebar, mm
nbar = 12  # Number of main reinforcement bars
nbar_side = 2  # Number of rebars at each side
cc = 40    # Concrete cover, mm
Ec = 30000 # MPa (concrete elastic modulus)
Es = 2e5   # MPa (steel elastic modulus)
tie = 10   # Tie/hoop diameter, mm (assumed)

# ======================
# Derived Values
# ======================
# Effective Depth (d)
deff = h - cc - tie - dbar/2
# Bar depth d'
dprime = cc + tie + dbar/2

# Compute beta1
if fc <= 28:
    beta1 = 0.85
else:
    # per ACI: beta1 = 0.85 - 0.05 * (fc - 28)/7, but not less than 0.65
    beta1 = max(0.85 - 0.05 * (fc - 28) / 7, 0.65)

# Tension steel area (each side)
Ast = (math.pi * dbar**2 / 4) * nbar_side
Asc = Ast  # same area at compression side, if symmetrical

# Concrete strain at crushing
ec = 0.0033  
# Steel yield strain
ey = fy / Es

# ================
# Global containers
# ================
nom_load = []
ult_load = []
nom_moment = []
ult_moment = []
eccentricity = []
phi_factor_list = []

# ======================
# Define helper functions
# ======================
def stress_strain(c):
    """
    Returns:
    a        : depth of equivalent rectangular stress block (beta1*c)
    es       : steel strain at tension side
    fs       : steel stress at tension side
    esprime  : steel strain at compression side
    fsprime  : steel stress at compression side
    """
    global a, es, fs, esprime, fsprime
    es = ec*(deff - c)/c
    fs = Es * es
    
    esprime = ec*(c - dprime)/c
    fsprime = Es * esprime
    
    a = beta1 * c
    return a, es, fs, esprime, fsprime

def strength_factor():
    """
    Compute phi (strength reduction factor) and classification 
    based on tension- or compression-controlled section.
    """
    global phi, classify
    # We use the global "es" and "ey" from above
    phi_calc = 0.64 + 0.25 * (es - ey) / (0.005 - ey)
    if phi_calc >= 0.90:
        phi, classify = 0.90, 'Tension-controlled'
    elif phi_calc <= 0.65:
        phi, classify = 0.90, 'Compression-controlled'
    else:
        phi, classify = phi_calc, 'Transition'
    return phi, classify

def force_moments():
    """
    Returns:
    ecc : eccentricity in mm
    Pn  : nominal axial load in kN
    Pu  : ultimate axial load in kN
    Mn  : nominal moment in kN-m
    Mu  : ultimate moment in kN-m
    """
    global Pn, Pu, Mn, Mu, ecc
    # Tension force in steel at tension side
    T = Ast * min(fy, fs) / 1000.0  # kN
    
    # Compression force by concrete
    Cc = 0.85 * fc * a * b / 1000.0 # kN
    
    # Steel in compression zone: 
    # (fy or fsprime) minus the portion already "taken" by the concrete (0.85*fc)
    Cs = Asc * (min(fy, fsprime) - 0.85*fc) / 1000.0
    
    Pn = Cc + Cs - T
    # phi is from strength_factor()
    Pu = phi * Pn
    
    # Eccentricity: sum of moments about T = 0
    ecc = (Cc*(deff - a/2) + Cs*(deff - dprime)) / Pn - (h/2 - dprime)
    # Convert moment to kN-m
    Mn = (Pn * ecc) / 1000.0
    Mu = phi * Mn
    return ecc, Pn, Pu, Mn, Mu

# ===========================
# Solve axial load & moment
# for a range of c values
# ===========================
# If you want c from 500 down to 80 in steps of -5:
c_values = np.arange(500, 80, -5)

for c in c_values:
    a, es, fs, esprime, fsprime = stress_strain(c)
    phi, classify = strength_factor()
    ecc, Pn, Pu, Mn, Mu = force_moments()
    
    # A filter, for instance, ignoring very large eccentricities
    if ecc <= 1.5 * h:  
        nom_load.append(round(Pn))
        ult_load.append(round(Pu))
        nom_moment.append(round(Mn))
        ult_moment.append(round(Mu))
        eccentricity.append(round(ecc))
        phi_factor_list.append(round(phi, 2))

# ============================
# Create a DataFrame of results
# ============================
data = {
    'ecc(mm)':     eccentricity, 
    'Pn(kN)':      nom_load,
    'Pu(kN)':      ult_load,
    'Mn(kN-m)':    nom_moment,
    'Mu(kN-m)':    ult_moment,
    'phi':         phi_factor_list
}
df = pd.DataFrame(data)
print("\nLast 10 rows of the P-M data:")
print(df.tail(10))

# ==========================
# Plot the P-M interaction
# ==========================
sns.set_style('darkgrid')
fig, ax = plt.subplots(figsize=(13, 7))
plt.rc('axes', 
       labelweight='bold', 
       labelsize='large', 
       titleweight='bold',
       titlesize=15)

# Scatter plot of nominal P-M points
ax = sns.scatterplot(x=nom_moment, y=nom_load, color='r', ax=ax)

ax.set_xlabel('Moment (M, kN-m)')
ax.set_ylabel('Axial Load (P, kN)')
plt.title('P-M Interaction Diagram')
plt.show()
```
